```python
pygame.init()
screen = pygame.display.set_mode((900, 750))
clock = pygame.time.Clock()

# load the trained machine learning model
model = tf.keras.models.load_model("models/model.h5")

def draw_sketch():
    sketch = [(0,0), (0,0)]
    # WHITE = (255, 255, 255)
    # BLACK = (0, 0, 0)
    #
    # mouse_position = (0, 0)
    # drawing = False
    # screen = pygame.display.set_mode((600, 800), 0, 32)
    # screen.fill(WHITE)
    # pygame.display.set_caption("ScratchBoard")
    #
    # while True:
    #     for event in pygame.event.get():
    #         if event.type == pygame.KEYDOWN:
    #             if event.key == pygame.K_ESCAPE:
    #                 pygame.quit()
    #                 return
    #             # if event.key == pygame.MOUSEBUTTONDOWN:
    #             #     pygame.mouse.get_rel()
    #             #     sketch.append(pygame.mouse.get_pos())
    #             # elif event.key == pygame.MOUSEBUTTONUP:
    #             #     pygame.mouse.get_rel()
    #
    #             elif event.type == pygame.MOUSEMOTION:
    #                 if drawing:
    #                     mouse_position = pygame.mouse.get_pos()
    #                     pygame.draw.line(screen, BLACK, mouse_position, mouse_position, 1)
    #             elif event.type == pygame.MOUSEBUTTONUP:
    #                 mouse_position = (0, 0)
    #                 drawing = False
    #             elif event.type == pygame.MOUSEBUTTONDOWN:
    #                 drawing = True
    #     screen.fill((255, 255, 255))
    #     pygame.draw.lines(screen, (0, 0, 0), False, sketch, 5)
    #     pygame.display.update()
    #



    pygame.init()

    WHITE = (255, 255, 255)
    BLACK = (0, 0, 0)

    mouse_position = (0, 0)
    drawing = False
    screen.fill(WHITE)
    pygame.display.set_caption("ScratchBoard")

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                # sys.exit()
                return
            if event.type == pygame.MOUSEMOTION:
                if drawing:
                    mouse_position = pygame.mouse.get_pos()
                    pygame.draw.line(screen, BLACK, mouse_position, mouse_position, 1)
                    sketch.append(pygame.mouse.get_pos())
            elif event.type == pygame.MOUSEBUTTONUP:
                mouse_position = (0, 0)
                drawing = False
                sketch.append(pygame.mouse.get_pos())
            elif event.type == pygame.MOUSEBUTTONDOWN:
                drawing = True
                sketch.append(pygame.mouse.get_pos())
            elif event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
                # sketch_data = np.array([np.array(sketch)])
                # sketch_data = sketch_data / 255.0
                #
                #
                # sketch_data_resized = cv2.resize(sketch_data, (28, 28), interpolation=cv2.INTER_AREA)
                # sketch_data_resized = sketch_data_resized.reshape((-1, 28, 28, 1))

                # font = pygame.font.Font(None, 36)
                # text = font.render(str(sketch_data_resized.shape), True, (0, 0, 0))
                # screen.blit(text, (50, 50))

                # sketch_data = sketch_data_resized.reshape((28, 28, 1))
                # final_input = sketch_data_resized / 255.0
                # use the machine learning model to make a prediction

                # img = cv2.imread("anvil.png")
                # reimage = cv2.resize(img, (28, 28), interpolation=cv2.INTER_AREA)
                # final_input = img / 255.0
                # resizedImage = np.reshape(img, (-1, 28, 28, 1))


                drawing_array = pygame.surfarray.array3d(pygame.transform.scale(sketch, (28, 28)))[:, :, 0]
                drawing_array = np.expand_dims(drawing_array, axis=2)
                drawing_array = drawing_array.astype('float32') / 255.0

                prediction = model.predict(drawing_array)
                # guess = np.argmax(prediction)

                font = pygame.font.Font(None, 36)
                text = font.render(str(prediction), True, (0, 0, 0))
                screen.blit(text, (50, 50))
                # pygame.display.update()

        pygame.display.update()

        # try:
        #     sketch
        # except NameError:
        #         font = pygame.font.Font(None, 36)
        #         text = font.render(str("sketch1"), True, (0, 0, 0))
        #         text = font.render(str(sketch), True, (0, 0, 0))
        #         screen.blit(text, (50, 50))
        # else:


        # predict what the user has drawn using the machine learning model
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                # preprocess the sketch data
                sketch_data = np.array([np.array(sketch)])
                sketch_data = sketch_data / 255.0
                # use the machine learning model to make a prediction
                prediction = model.predict(sketch_data)
                guess = np.argmax(prediction)
                # display the AI's guess on the screen
                font = pygame.font.Font(None, 36)
                text = font.render(str(guess), True, (0, 0, 0))
                screen.blit(text, (50, 50))
                pygame.display.update()
if __name__ == "__main__":
    draw_sketch()
```
